# Using Matplotlib to define the company sales,revenue etc

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMAZON','MICROSOFT','FACEBOOK']
revenue=[9033,1326,8239,2327]
```

```python
xpos = np.arange(len(company))
xpos
```

# Simple bar chart representation

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.xlabel("Company")
plt.title('US Technology Stocks')
plt.legend()
```

# Showing the profit of the company

```python
profit=[402,354,457,456]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")
plt.xticks(xpos,company)
plt.ylabel("Profit(Bln)")
plt.xlabel("Company")
plt.title('US Technology Stocks')
plt.legend()
```
